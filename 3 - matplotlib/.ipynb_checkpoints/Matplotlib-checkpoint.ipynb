```python
unis = pd.read_csv('data/uk_universities.csv', nrows=50)
unis.head()
unis.columns
```

```python
#plt.plot(y)
plt.plot(unis['World_rank'] , unis['Student_enrollment'])
#should there be null values in the data, use unis.fillna(0, inplace=True)
```

```python
plt.figure(figsize=(10,6))
plt.plot( unis['World_rank'], unis['Student_enrollment'])
plt.show() #it hides the first data line row
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,2,2)
plt.plot(unis['World_rank'], unis['Student_enrollment'], color='r', linestyle='--', 
         linewidth=2, marker='o', markersize=10, markerfacecolor='b', markeredgecolor='b', markeredgewidth=1)
plt.xlabel('Rank')
plt.ylabel('Enrollment num')
plt.title('Yes')
```

```python
fig = plt.figure(figsize=(10,3))
ax1= fig.add_axes([0,0,1,1])
ax2= fig.add_axes([0.1,0.1,0.2,0.4])
ax1.plot(unis['World_rank'], unis['Student_enrollment'], color='r', linestyle='--', 
         linewidth=2, marker='o', markersize=10, markerfacecolor='b', markeredgecolor='b', markeredgewidth=1)
ax1.set_xlabel('Rank') #not xlabel
ax1.set_ylabel('Enrollment num')

ax2.set_xlim(20,25)
ax2.set_ylim(4,8)
ax2.plot(unis['World_rank'], unis['Student_enrollment'], color='r', linestyle='--', 
         linewidth=2, marker='o', markersize=10, markerfacecolor='b', markeredgecolor='b', markeredgewidth=1)
ax2.set_xlabel('Rank') #not xlabel
ax2.set_ylabel('Enrollment num')

#this looks so messed up but basically what this does is that it creates a little zoomed in version of a plot
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.plot(unis['World_rank'], unis['Student_enrollment'], linestyle='--', color='r', label='Rating')

plt.xlim(0,500)
plt.legend()
plt.grid()
plt.xticks(np.arange(0,30,4))
```
