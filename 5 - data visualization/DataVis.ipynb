```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris') #its from github
```

```python
iris.head()
```

```python
iris.describe()
```

```python
plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams['figure.dpi'] = 300
iris.plot()
```

```python
iris.plot(logy=True, logx=True)
plt.title('Iris dataset plot')
plt.xlabel('Samples (log scale)')
plt.ylabel('Length/Width')
plt.show()
```
