```python
#!pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from numpy.random import randint

%matplotlib inline
```

```python
x = np.linspace(1,10,20) #in hte plot the max is 10 on the x axis, min 1, and there are 20 steps
y = randint(1,50,20) #the steps must match
```

```python
x.size, y.size
```

```python
#help(plt)

plt.plot(x, y)

plt.figure() #this creates a new figure, otherwise the two graphs will be merged into one figure
plt.plot(y,x)
```
