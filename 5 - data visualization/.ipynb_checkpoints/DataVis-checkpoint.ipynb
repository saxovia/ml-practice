```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris') #its from github
```

```python
iris.head()
```

```python
iris.describe()
```

```python
plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams['figure.dpi'] = 300
iris.plot()
```

```python
iris.plot(logy=True, logx=True)
plt.title('Iris dataset plot')
plt.xlabel('Samples (log scale)')
plt.ylabel('Length/Width')
plt.show()
```

```python
x = iris[['sepal_width', 'sepal_length']]
y = iris[['petal_width', 'petal_length']]
```

```python
ax = x.plot()
ax2 = y.plot(ax = ax, secondary_y=True)

ax.set_ylabel('Sepal (Width and Length)')
ax2.set_ylabel('Petal (Width and Lenght)')

ax.set_xlabel('#Samples')
ax.set_title('Secondary Axis Plot')
plt.show()
```

```python
df = iris.iloc[0][0:4]
```

```python
df.plot(kind='bar')

df.plot.bar()
df.plot.bar(color='red', fontsize=5, title=':3', ylabel='length/width', legend=True, 
            ylim=(0,7), grid=True, rot=30) #rot as in rotate labels
```

```python
df = iris.head()
df.plot.bar(fontsize=5, title=':3', ylabel='length/width', legend=True, 
            ylim=(0,12), stacked=True, color= ['red', 'orange', 'purple', 'blue'])
plt.legend(bbox_to_anchor=(1.5,1))
plt.show()
```
